```python
import requests
from bs4 import BeautifulSoup
```

```python
response=requests.get('https://www.screener.in/screens/357649/all-listed-companies/?page=2')
soup=BeautifulSoup(response.content,'html.parser')
a_tags=soup.find_all('a')
```

```python
all_companies=[]
i=1
for page_no in range(1,173):

  response=requests.get(f'https://www.screener.in/screens/357649/all-listed-companies/?page={page_no}')
  soup=BeautifulSoup(response.content,'html.parser')
  a_tags=soup.find_all('a')



  for a in a_tags:
    if 'company' in a.get('href') :
      all_companies.append(a.get('href'))

      print('{i} {name}'.format(i=i,name=a.get('href')))
      i+=1
      # print(a.get('href'))
print(i)
```

```python
len(all_companies)
```

```python
import numpy as np
import pandas as pd
```

```python
arr=np.array(all_companies)
```

```python
df=pd.DataFrame(arr,columns=['company_url'])
```

```python
df.head(3)
```

```python
df['company name']=df['company_url'].apply(lambda x:x.split('/')[2])
```

```python
df['screener url']=df['company_url'].apply(lambda x:'https://www.screener.in'+x)
```

```python
df_final=df[['company name','company_url','screener url']]
```
